```python
attributes = ['crossing', 'finishing', 'heading_accuracy', 'short_passing',
              'volleys', 'dribbling','curve', 'free_kick_accuracy', 
              'long_passing', 'ball_control', 'acceleration', 
              'sprint_speed', 'agility', 'reactions', 'balance', 
              'shot_power', 'jumping', 'stamina', 'strength', 'long_shots',
              'aggression', 'interceptions', 'positioning', 'vision',
              'penalties', 'composure', 'marking', 'standing_tackle', 
              'sliding_tackle', 'gk_diving', 'gk_handling', 'gk_kicking',
              'gk_positioning', 'gk_reflexes'
             ]

positions = ['rs', 'rw', 'rf', 'ram', 'rcm', 'rm', 'rdm', 'rcb', 'rb', 
             'rwb', 'st', 'lw', 'cf', 'cam', 'cm', 'lm', 'cdm', 'cb', 'lb',
             'lwb', 'ls', 'lf', 'lam', 'lcm', 'ldm', 'lcb', 'gk'
            ]

cols_pref = [x for x in df.columns if 'prefers' in x and 'gk' not in x]

min_pref_pos = min(df.loc[:, cols_pref].sum(axis=1))
max_pres_pos = max(df.loc[:, cols_pref].sum(axis=1))

print('minimum of preferred positions of a player is {0}.'.format(min_pref_pos))
print('maximum of prefereed positions of a player is {0}.'.format(max_pres_pos))
```
