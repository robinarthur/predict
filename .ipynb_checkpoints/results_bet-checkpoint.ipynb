```python
#Gets all the statistics related to draw betting strategy

columns_req = ['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']

odds_statistics_14 = raw_data_14[columns_req]
odds_statistics_15 = raw_data_15[columns_req]
odds_statistics_16 = raw_data_16[columns_req]
odds_statistics_17 = raw_data_17[columns_req]
odds_statistics_18 = raw_data_18[columns_req]
```

```python

odds_statistics_14['RESULT'] = odds_statistics_14['FTHG'].map(str) + ' : ' + odds_statistics_14['FTAG'].map(str)
odds_statistics_15['RESULT'] = odds_statistics_15['FTHG'].map(str) + ' : ' + odds_statistics_15['FTAG'].map(str)
odds_statistics_16['RESULT'] = odds_statistics_16['FTHG'].map(str) + ' : ' + odds_statistics_16['FTAG'].map(str)
odds_statistics_17['RESULT'] = odds_statistics_17['FTHG'].map(str) + ' : ' + odds_statistics_17['FTAG'].map(str)
odds_statistics_18['RESULT'] = odds_statistics_18['FTHG'].map(str) + ' : ' + odds_statistics_18['FTAG'].map(str)



result14 = odds_statistics_14['RESULT'].value_counts()
result15 = odds_statistics_15['RESULT'].value_counts()
result16 = odds_statistics_16['RESULT'].value_counts()
result17 = odds_statistics_17['RESULT'].value_counts()
result18 = odds_statistics_18['RESULT'].value_counts()

#print('results of 2014:\n', result14)
#print('results of 2015:\n', result15)
#print('results of 2016:\n', result16)
#print('results of 2017:\n', result17)
#print('results of 2018:\n', result18)

s = []

for i14, v14 in result14.items():
    for i15, v15 in result15.items():
        for i16, v16 in result16.items():
            for i17, v17 in result17.items():
                for i18, v18 in result18.items():
                    if i14 == i15 == i16 == i17 == i18:
                        idurch = (v14 + v15 + v16 + v17 + v18) / 5
                        s.append((i14, idurch))


for i, v in s:
    print('index: ', i, 'value: ', v)
    
print(s)
    
```
